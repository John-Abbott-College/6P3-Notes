## Functions

Minimal definition of a function:

```python
# An empty function
def function():
	pass
```

`pass` is used as a set of empty `{}` to satisfy the interpreter.


Example with data-types for arguments and return value

```python
# Receives a list of numbers and returns the average
def average(grades: list)-> float:
    
    total = sum(grades)  # sum is a built-in global function
    entries = len(grades)
    return total/entries

grades = [78.5, 81, 100, 71.25, 60]
print( average(grades) )
```

### Optional arguments

```python
def play_game(game, players=1):
	if players == 1:
		print(f'Playing {game} by myself')
	else:
		print(f'{game} is a lot more fun with more people')		

play_game("tic-tac-toe")
```

### Keyword & Positional Arguments

Function arguments can be passed as:

- Positional arguments _(the standard)_
- Named arguments
- Combination of the above


```python
from datetime import date

def borrow_item(item, borrow_date, return_date):
    print( f"You are borrowing my {item} from {borrow_date} to {return_date}")
    
now = date.today()
next_week = now.replace(day = now.day + 7);

# positional, positional, keyword
borrow_item("camera", now, return_date = next_week)
```

**Note**
- Positional arguments must always come before keyword arguments.

### `*args` and `**kwargs`

Functions can take an arbitrary number or arguments.

Define the argument `*args` to:

- Take unlimited positional arguments;
- Arguments are available inside the function as a **Tupple**.

```python
def place_order(item, *args):
    print( f"You are ordering a {item}")
    for order in args:
        print(order)

# Order a single item
place_order("hotdog")

# Order lots
place_order("pizza", "fries", "coffee", "chocolate", "cola")
```

Define the argument `**kwargs` to:

- Receive an unlimited number of keyword arguments;
- These arguments are available inside the function as a **Dictionary**.

```python
def place_order_extra(item, **kwargs):
    print( f"You are ordering a {item}")
    for key in kwargs:
        print(f"As a {key}: {kwargs[key]}")

# Order a single item
place_order_extra("hotdog")

# Order lots
place_order_extra("pizza", side="fries", drink="coffee", desert="chocolate")
```

**Note**
- The variable names `args` and `kwargs` are simply a convention.
- It is the `*` and `**` operators that unpack the arguments.

### Constructor & Attributes

The method `__init__()` is the class "constructor".

- `self` is the first constructor argument and **is mandatory**.
    - It refers to the current class instance.

```python
class Person:
    # constructor
    def __init__(self, name:str, age:int):
        self.name = name
        self.age = age

    # method
    def greeting(self):
        print(f"Hi, I'm {self.name}")

me = Person("Mauricio", 36)
print( me )

# dictionary representation of the object
print( me.__dict__ )
```

**Note**

- Printing the object does not results in a "nice print"
- The method `__repr__` or `__str__` define the string representation of the object.

```python
class Pet:
    def __init__(self, name:str, age:int, species:str):
        self.name = name
        self.age = age
        self.species = species

    # Defined the string representation of the object
    def __repr__(self):
        return f"Pet ({self.name} the {self.species} is {self.age})"

my_cat = Pet("Jack", 4, "cat")
my_dog = Pet("Loki", 3, "dog")

print( my_cat )
print( my_dog )
```

### Class Inheritance

Inherit from a parent class by including the parent class in its declaration:

```python
class Student(Person):
    def __init__(self, name, age, student_number):
        super().__init__(name, age)    # use the parent's constructor
        self.student_number= student_number
        
    def study(self):
        print("üò≠")

jimmy = Student("jimmy", 20, 123456)
jimmy.greeting()   # inherited from parent class
jimmy.study()
```

### Private Class Members

In Python has no private class attributes, **only conventions**.





The convention is based on prefixes onthe class member name:

- Prefixed with underscore (e.g. `_spam`):
    - Treat it as private. _(aka. don't touch this)_

- Prefixed with double-underscore (e.g. `__salary`):
    - Treat it as private and cannot be accessed by name alone. _(aka. can't touch this)_

```python
class Teacher(Person):
    def __init__(self, name: str, age: int, email: str, salary: int):
         super().__init__(name, age)
         self._email = email
         self.__salary = salary

    def __procrastinate(self):
        print( "YouTube... Yeah!")

vik = Teacher("Vik", 33, "vik@jac.qc.ca", 99999)

print( vik._email )   # supposed to treat it as private!

print( vik.__salary )  # can't touch this
vik.__procrastinate()
```

The double underscore convention changes the attribute name with `_className__` to make it difficult to access by "mistake". However, it's still there. This technique is called _name mangling_.

```python
# Accessing "private" attribute
print( vik._Teacher__salary )
vik._Teacher__procrastinate()
```
